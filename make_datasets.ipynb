````python
curse_df = pd.read_csv('Curse-detection-data/dataset.txt', sep='|', names=['문장', 'hate'])
````

````python
curse_df['clean'] = curse_df['hate'].apply(lambda x: 0 if x == 1 else 1)
curse_df['labels'] = curse_df[['hate', 'clean']].values.tolist()
curse_df
````

````python
!git clone https://github.com/kocohub/korean-hate-speech.git
````

````python
korean_hate_train_df = pd.read_csv('korean-hate-speech/train.tsv', sep='\t')
korean_hate_dev_df = pd.read_csv('korean-hate-speech/dev.tsv', sep='\t')
korean_hate_df = pd.concat([korean_hate_train_df, korean_hate_dev_df]).reset_index(drop=True)
````

````python
korean_hate_df['hate'] = korean_hate_df['hate'].apply(lambda x: 0 if x == 'none' else 1)
korean_hate_df['clean'] = korean_hate_df['hate'].apply(lambda x: 1 if x == 0 else 0)
korean_hate_df['labels'] = korean_hate_df[['hate', 'clean']].values.tolist()
korean_hate_df = korean_hate_df.rename(columns={
    'comments': '문장'
})
korean_hate_df = korean_hate_df[['문장', 'hate', 'clean', 'labels']]
````

````python
korean_hate_df
````

````python
!git clone https://github.com/seongwoong/korean-malicious-comments-dataset.git
````

````python
korean_malicious_comments_df = pd.read_csv('korean-malicious-comments-dataset/Dataset.csv', sep='\t')
korean_malicious_comments_df = korean_malicious_comments_df.dropna()
korean_malicious_comments_df['lable'] = korean_malicious_comments_df['lable'].astype(int)
````

````python
korean_malicious_comments_df = korean_malicious_comments_df.rename(columns={'lable': 'clean'})
korean_malicious_comments_df['hate'] = korean_malicious_comments_df['clean'].apply(lambda x: 1 if x == 0 else 0)
korean_malicious_comments_df['labels'] = korean_malicious_comments_df[['hate', 'clean']].values.tolist()
korean_malicious_comments_df = korean_malicious_comments_df.rename(columns={
    'content': '문장'
})
korean_malicious_comments_df = korean_malicious_comments_df[['문장', 'hate', 'clean', 'labels']]
````

````python
korean_malicious_comments_df
````

````python
data_list = [unsmile_df, curse_df, korean_hate_df, korean_malicious_comments_df]
df = pd.concat(data_list).reset_index(drop=True)
````

````python
df
````

````python
df = df[['문장', 'hate', 'clean']].drop_duplicates()
df['labels'] = df[['hate', 'clean']].values.tolist()
````

````python
df[['문장', 'hate']]
````

````python
from sklearn.model_selection import train_test_split
train, valid = train_test_split(df[['문장', 'hate']], test_size=0.2)
````

````python
train.to_parquet('train.parquet')
valid.to_parquet('valid.parquet')
````
